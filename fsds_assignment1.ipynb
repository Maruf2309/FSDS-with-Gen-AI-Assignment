**Assignment 1: Temperature Check**

Problem:

Write a program that takes the current temperature as input and prints:

* "It's too cold!" if below 10°C
* "It's cool outside" if between 10°C and 25°C
* "It's hot outside!" if above 25°C



```python
##Solution 1
temp = int(input("Enter current temperature: "))

if temp < 10 :
    print("Its too cold!")

elif 10 <= temp <= 25:
    print("Its cool outside")

else:
    print("Its hot outside!")
```

**Assignment 2: ATM Withdrawal System**

Problem:

Ask the user for their account balance and the amount they want to withdraw:

- If the withdrawal amount is greater than the balance → print "Insufficient
Balance"
- If the withdrawal amount is less than or equal to balance → print "Transaction
Successful" and show the remaining balance.

```python
## Solution 2:

balance  = int(input("Enter your balance: "))
withdrawal = int(input("Enter your withdrawal amount: "))



if withdrawal > balance:
    print("Insufficient Balance!")

else:
    balance -= withdrawal
    print("Transaction is successfull!")
    print("Remaining Balance:", balance)
```

**Assignment 3: Student Grade Calculator**

Problem:

Input marks `(0–100)` and print grade based on:

- 90 - 100: A+
- 80 - 89: A
- 70 - 79: B
- 60 - 69: C
- Below 60: Fail

```python
## Solution 3: 

marks = int(input("Enter Student Marks (0-100): "))

if marks < 0 or marks > 100:
    print("Invalid marks! Please enter a value between 0 and 100.")
elif 90 <= marks <= 100:
    print("Student Grade is A+")
elif 80 <= marks <= 89:
    print("Student Grade is A")
elif 70 <= marks <= 79:
    print("Student Grade is B")
elif 60 <= marks <= 69:
    print("Student Grade is C")
else:
    print("Fail!")
```

**Assignment 5: Restaurant Bill with Discount**

Problem:

Take total bill amount from the user:

- If the bill is above 1000 → apply 10% discount
- Otherwise → no discount 
- Then print the final bill amount

```python
# Solution 5

bill = float(input("Enter total bill amount: "))

if bill > 1000:
    discount = bill * .10
    final_bill = bill - discount
    print("10% Discont Applied for the bill!")
else:
    final_bill = bill 
    print("No Discount applied!")

print(f"Final bill amount: {final_bill}")
```

**Assignment 8: Leap Year Checker**

Problem:

Take a year as input and check:

- if it is divisible by 400 → Leap Year
- Else if divisible by 4 but not by 100 → Leap Year
- Otherwise → Not a Leap Year

```python
# Assignment 8:

year = int(input("Enter a year: "))

if year % 400 == 0:
    print(f"{year} is a Leap Year")
elif year % 4 == 0 and year % 100 != 0: # divisible by 4 but not divisible by 100 → leap year.
    print(f"{year} is a Leap Year")
else:
    print(f"{year} is Not a Leap Year")
```

**Assignment 10: Electricity Bill Calculator**

Problem:

Input number of units consumed and calculate bill:

- Up to 100 units → 5 BDT per unit
- 101–200 units → 7 BDT per unit
- Above 200 units → 10 BDT per unit
- Print total payable amount

```python
## Solution 10

units = int(input("Enter the number of units consumed: "))

if units <= 100:
    bill = units * 5
elif 101 <= units <= 200:
    bill = units * 7
else:  # units > 200
    bill = units * 10

print(f"Total Payable Amount: {bill} BDT")
```
