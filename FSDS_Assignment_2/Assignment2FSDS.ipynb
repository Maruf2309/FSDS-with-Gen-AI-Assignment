**Task1: Email Validator (String Filtering + Loops)**

Real-world use: Used in signup forms or data cleaning tasks.

**Task:**

- Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.

Example:

```python
Input: hello@inceptionbd.com
Output: Valid Email
```


````python
# Solution 1:

email = input("Enter email: ").strip()

# Check exactly one '@'
if email.count("@") == 1:

    # Split into local and domain
    local, domain = email.split("@")

    # Both must be non-empty and domain must end with .com or .org
    if local != "" and domain != "" and (domain.endswith(".com") or domain.endswith(".org")):
        print("Valid Email")
    else:
        print("Invalid Email")

else:
    print("Invalid Email")
````

# 3. Password Strength Checker

Real-world use: Used in user authentication systems.

**Task:**

Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.

Example:
```python
Input: "Hello@123"
Output: Strong Password
```


**Solution1**

‚úÖ Strong Password Checker (Uppercase + Lowercase + Digit + Special Char)

````python
# Without using any function

password = input("Enter Password: ")

# Four flag to check different conditions 

has_upper = False  # becomes True if password has any uppercase letter (A‚ÄìZ)
has_lower = False
has_digit = False
has_special = False

special_chars = "!@#$^&*()-_=+{}[]|;:,.<>?/"

# Check each character

for ch in password:  # This loop will check characters one by one (H e l l o @ 1 2 3)
    if ch.isupper():
        has_upper = True
    elif ch.islower():
        has_lower = True
    elif ch.isdigit():
        has_digit = True
    elif ch in special_chars:
        has_special = True

# Final Check
if has_upper and has_lower and has_digit and has_special:
    print("Strong Pasword")
else:
    print("Week Password")
````

# 4. Username Generator

**üíº Real-world use: Automatically create usernames for users.**

Task:

- Take user‚Äôs full name and generate a short username.
- Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).

Example:
```python
Input: "Mamun Malitha"
Output: mamitha57

```

````python
import random

# Ask user for full name
full_name = input("Enter your full  name: ").strip().lower()
parts = full_name.split()

first_name = parts[0]
last_name = parts[-1]

## Extact first 3 latters using loop
first3 = ""
count = 0
for ch in first_name:
    if count <3:
        first3 += ch
        count += 1

## Extract last 3 letters using loop
last3 = ""
length = 0
for _ in last_name:
    length +=1


## Start index = lenght - 3
index = 0
for ch in last_name:
    if index >= length - 3:
        last3 += ch
    index += 1

# Generate Random 2 digit number
num = random.randint(10, 99)

# Final username
username = first3 + last3 + str(num)
print("Generated User Name:", username)
````

# 5. SMS Word Counter

**üíº Real-world use: Used in messaging apps or social media post limit check.**

Task:

Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

Example:
```python
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit
```

````python
message = input("Enter your message: ")

# Count Characters
char_count = 0
for ch in message: 
    char_count += 1

# Count words (manual split using loop)
word_count = 0
in_word = False

for ch in message:
    if ch!= " " and not in_word: 
        word_count += 1
        in_word = True 

    elif ch == " ":
        in_word = False

# Check SMS limit (160)

if char_count > 160:
    status = "Exceeds Limit"
else:
    status = "Within Limit"

# Output
print("Words =", word_count)
print("Characters =", char_count)
print("Status =", status)
````

# 6. Remove Duplicates from Contact List

**üíº Real-world use: Data cleaning in CRM or user databases.**

Task:

Given a list of names (comma separated), remove duplicates.
Example:
```python
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"
```

````python
names = "Rahim, Karim, Rahim, Sultana"

unique = []
for n in names.split(", "):
    if n not in unique:
        unique.append(n)

output = ", ".join(unique)
print(output) 
````

# 7. Detect Suspicious Words in Text (Moderation)

**üíº Real-world use: Used in chat moderation or email filtering.**

Task:

If any ‚Äúbanned‚Äù words like `['spam', 'scam', 'fake']` are found in a text, print ‚ÄúWarning‚Äù.

Example:
```python
Input: "This is a scam offer"
Output: Warning: Suspicious content detected
```

‚úÖ Detect Suspicious Words (Moderation System)

````python
def detect_suspicious(text):
    banned_words = ['spam', 'scam', 'fake']

    # Convert text to lowercase for case-insensitive checking
    text_lower = text.lower()

    # Check if any banned word exists in the text
    for word in banned_words:
        if word in text_lower:
            return "Warning: Suspicious content detected"

    return "Clean: No suspicious content found"


# --- User Input & Output ---
text = input("Enter your text: ")

result = detect_suspicious(text)
print(result)

````

# 10. Word Reversal for Data Encryption

**üíº Real-world use: Simple text obfuscation or playful encoding.** 

Task: Reverse each word in a sentence while keeping word order the same. 
Example: 
```python
Input: "AI is fun" 
Output: "IA si nuf"
```

````python
# ‚úÖ Python Code: Reverse Each Word
sentence = "AI is fun"

# Split the sentence into words
words = sentence.split()

# Reverse each word
reversed_words = [word[::-1] for word in words]

# Join back into a sentence
output = " ".join(reversed_words)

print(output)   # IA si nuf
````
